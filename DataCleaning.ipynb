```python
#open replays_2024-02-12.txt from data folder

import json

#open file
with open('data/replays_2024-02-12.txt', 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
d = data[0][0]['log']
#split D by each \n
d = d.split('\n')
d
```

```python
# Initialize an empty list to hold the events
events = []

# Loop over each line in the log
for line in d:
    # Split the line into parts based on the '|' character
    parts = line.split('|')

    # Remove any empty strings from the list of parts
    parts = [part for part in parts if part]

    # Skip this iteration if parts is empty
    if not parts:
        continue

    # Create a dictionary for the event
    event = {'event_type': parts[0]}

    # Add the remaining parts to the dictionary with keys 'detail_1', 'detail_2', etc.
    for i, part in enumerate(parts[1:], start=1):
        event[f'detail_{i}'] = part

    # Add the event to the list of events
    events.append(event)

# Now 'events' is a list of dictionaries, where each dictionary represents an event
```

```python
#remove event_type of j, teamsize, gametype, gen, tier, rule
events = [event for event in events if event['event_type'] not in ['j', 'teamsize', 'gametype', 'gen', 'tier', 'rule']]
events
```

```python
#get events of event_type 't:'
team_events = [event for event in events if event['event_type'] == 't:']
t = team_events[0]['detail_1']

#turn t ('1707748164') into a timestamp
import datetime
timestamp = datetime.datetime.fromtimestamp(int(t))
timestamp.strftime('%Y-%m-%d %H:%M:%S')
```

```python
formattedEvents = {}

turns = []
current_turn = []

#get players
for event in events:
    if event['event_type'] == 'player':
        if event['detail_1'] == 'p1':
            formattedEvents['player_1'] = event['detail_2']
        elif event['detail_1'] == 'p2':
            formattedEvents['player_2'] = event['detail_2']
    #get starting pokemon (first instance of switch)
    if event['event_type'] == 'switch':
        ppSplit = event['detail_1'].split(': ')
        if ppSplit[0] == 'p1a':
            formattedEvents['p1_start_pokemon'] = ppSplit[1]
        elif ppSplit[0] == 'p2a':
            formattedEvents['p2_start_pokemon'] = ppSplit[1]
    
#remove events of event_type 'player'
events = [event for event in events if event['event_type'] != 'player']

#remove everything before the first event_type of 'turn'
for event in events:
    if event['event_type'] == 'turn':
        events = events[events.index(event):]
        break



for event in events:
    #if turn append last turn if not empty
    firstMove = {}
    secondMove = {}
    if event['event_type'] == 'turn':
        if current_turn:
            turns.append(current_turn)
            current_turn = [{'turn': event['detail_1']}] # Start a new turn
        else:
            current_turn.append({'turn': event['detail_1']})
    elif event['event_type'] == 't:':
        current_turn.append({'timestamp': event['detail_1']})
    elif event['event_type'] == 'move':
        moveSplit1 = event['detail_1'].split(': ')
        moveSplit2 = event['detail_3'].split(': ')
        if firstMove is {}:
            firstMove = {'usingPlayer': moveSplit1[0] , 'usingPokemon': moveSplit1[1], 'action': 'move', 'move': event['detail_2'],
                         'targetPlayer': (moveSplit2[0] if 0 < len(moveSplit2) else None), 'targetPokemon': (moveSplit2[1] if 1 < len(moveSplit2) else None)}
            current_turn.append({'firstMove': firstMove})
        else:
            secondMove = {'usingPlayer': moveSplit1[0] , 'usingPokemon': moveSplit1[1], 'action': 'move', 'move': event['detail_2'],
                         'targetPlayer': (moveSplit2[0] if 0 < len(moveSplit2) else None), 'targetPokemon': (moveSplit2[1] if 1 < len(moveSplit2) else None)}
            current_turn.append({'secondMove': secondMove})
    elif event['event_type'] == 'switch':
        switchSplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'switch', 'player': switchSplit[0], 'pokemon': switchSplit[1]})
    elif event['event_type'] == '-ability':
        abilitySplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'ability', 'player': abilitySplit[0], 'pokemon': abilitySplit[1], 'ability': event['detail_2']})
    elif event['event_type'] == '-item':
        itemSplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'item', 'player': itemSplit[0], 'pokemon': itemSplit[1], 'item': event['detail_2']})
    elif event['event_type'] == '-damage':
        damageSplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'damage', 'player': damageSplit[0], 'pokemon': damageSplit[1], 'damage': event['detail_2']})
    elif event['event_type'] == '-heal':
        healSplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'heal', 'player': healSplit[0], 'pokemon': healSplit[1], 'heal': event['detail_2']})
    elif event['event_type'] == '-status':
        statusSplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'status', 'player': statusSplit[0], 'pokemon': statusSplit[1], 'status': event['detail_2']})
    elif event['event_type'] == '-boost':
        boostSplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'boost', 'player': boostSplit[0], 'pokemon': boostSplit[1], 'boost': event['detail_2']})
    elif event['event_type'] == '-unboost':
        unboostSplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'unboost', 'player': unboostSplit[0], 'pokemon': unboostSplit[1], 'unboost': event['detail_2']})
    elif event['event_type'] == '-field':
        current_turn.append({'action': 'field', 'field': event['detail_1']})
    elif event['event_type'] == '-weather':
        current_turn.append({'action': 'weather', 'weather': event['detail_1']})
    elif event['event_type'] == 'upkeep':
        current_turn.append({'action': 'upkeep'})
    elif event['event_type'] == 'faint':
        faintSplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'faint', 'player': faintSplit[0], 'pokemon': faintSplit[1]})
    elif event['event_type'] == 'win':
        current_turn.append({'action': 'win', 'player': event['detail_1']})
    elif event['event_type'] == 'tie':
        current_turn.append({'action': 'tie'})
    elif event['event_type'] == 'immune':
        immuneSplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'immune', 'player': immuneSplit[0], 'pokemon': immuneSplit[1], 'move': event['detail_2']})
    elif event['event_type'] == '-superEffective':
        superEffectiveSplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'superEffective', 'player': superEffectiveSplit[0], 'pokemon': superEffectiveSplit[1], 'move': event['detail_2']})
    elif event['event_type'] == '-resisted':
        resistedSplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'resisted', 'player': resistedSplit[0], 'pokemon': resistedSplit[1]})
    elif event['event_type'] == '-singleturn':
        singleTurnSplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'singleTurn', 'player': singleTurnSplit[0], 'pokemon': singleTurnSplit[1], 'move': event['detail_2']})
    elif event['event_type'] == '-terastallize':
        terastallizeSplit = event['detail_1'].split(': ')
        current_turn.append({'action': 'terastallize', 'player': terastallizeSplit[0], 'pokemon': terastallizeSplit[1], 'move': event['detail_2']})
    else:
        #print(event)
        current_turn.append({'action': 'other', 'other': event['detail_1']})

        


# Append the last turn if it's not empty
if current_turn:
    turns.append(current_turn)

turns
```
